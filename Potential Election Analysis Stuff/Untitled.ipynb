```python
file_one = pd.read_csv("CD Demographics.csv")

file_one ["White"] = file_one ["White"] / file_one ["Total population"]
file_one ["Black or African American"] = file_one ["Black or African American"] / file_one ["Total population"]
file_one ["Asian"] = file_one ["Asian"] / file_one ["Total population"]
file_one ["Hispanic or Latino (of any race)"] = file_one ["Hispanic or Latino (of any race)"] / file_one ["Total population"]
file_one.head()
```

```python
file_one.iloc[:,0:15]
```

```python
list_dfs = []
for i in range(len(file_one["District"])):
    cd_df = file_one.iloc[i,:]
    new_df = pd.DataFrame(cd_df)
    new_df = new_df.reset_index()
    new_df = new_df.rename(columns={"index": "Category", 1: "District Characteristics"})
    list_dfs.append(new_df)
list_dfs[0]
```

```python
html_list = []
for df in list_dfs:
    new_df = df.to_html()
    df.to_html(f'{df.iloc[0,1]}_demographics.html',table_id= {df.iloc[0,1]})
    html_list.append(new_df)
html_list[0]
```
