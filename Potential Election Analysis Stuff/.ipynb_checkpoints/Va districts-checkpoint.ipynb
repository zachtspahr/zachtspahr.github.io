```python
file_one = pd.read_csv("house_data.csv")
file_one = file_one.drop(columns = ["pure_urban.1","rural_suburban_mix","Democrat Strength Rating","sparse_suburban","urban_suburban_mix","flipped_seat.1","Unnamed: 44","Unnamed: 45","Unnamed: 46","Unnamed: 47","Unnamed: 48","dense_suburban","pure_urban","Type of District","Unnamed: 35","Final?"])
file_one ["CD"] = file_one["State Abbreviation"] + "-" + file_one["CD#"].map(str)
file_one ["Last Two Pres Winners"] = file_one["2012 Winner"] + "/" + file_one["2016 winner"]
file_one = file_one.set_index('CD')
#file_one["Dem %"] = file_one["Dem %"].str.replace('%', '')
#file_one["GOP %"] = file_one["GOP %"].str.replace('%', '')
#file_one["Other %"] = file_one["Other %"].str.replace('%', '')
#file_one["Dem Margin"] = file_one["Dem Margin"].str.replace('%', '')
#file_one["2016 Clinton Margin"] = file_one["2016 Clinton Margin"].str.replace('%', '')
#file_one["Swing vs. 2016 Prez"] = file_one["Swing vs. 2016 Prez"].str.replace('%', '')
file_one.head()
states = list(file_one['State'].unique())
```

```python
va = file_one[file_one["State"]=="Virginia"]
va = va.rename(columns={"2018 Cook PVI Score": "PVI"})
va = va.sort_values(by=['CD#'])

va = va.reset_index()
#va['Dem %'] = va['Dem %'].astype(float)
va
#va["PVI"] = va["PVI"].astype(str)
#va["PVI"] = va["PVI"].str.replace('-', 'R+')


va['flipped_seat'] = va['flipped_seat'].fillna(0)
va= va.sort_values('CD#')
va ['PVI_int'] = va['PVI']
va = va.reset_index()
va = va.drop(columns = ["index"])   



midterm_dem_vote = (va["Dem %"])

#va.iloc[0,5]

va
```

```python


va['PVI'] = va ['PVI'].astype(int)
pvi = list(va['PVI'])
pvi
new_pvi_list = []

for i in pvi:
    if i <=0 :
        i = i* (-1)
        i = str(i)
        i = "R+" + i
        print (i)
        new_pvi_list.append(i)
    else:
        i= str(i)
        i = "D+" + i
        print(i)
        new_pvi_list.append(i)

va['PVI'] = new_pvi_list
va
```

```python
list_dfs = []
for i in range(len(va["CD#"])):
    cd_df = va.iloc[i,:]
    new_df = pd.DataFrame(cd_df)
    new_df = new_df.reset_index()
    new_df = new_df.rename(columns={"index": "Category", 1: "District Characteristics"})
    list_dfs.append(new_df)
list_dfs[1].head()

```

```python
html_list = []
for df in list_dfs:
    new_df = df.to_html()
    df.to_html(f'{df.iloc[0,1]}.html',table_id= {df.iloc[0,1]})
    html_list.append(new_df)
html_list[0]
```
